```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("x-Label")
plt.ylabel("y-Label")
plt.show()
```

```python
plt.plot(x,y,color="g",linestyle=':',linewidth=2)
plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.plot(x,y2,color='r',linestyle='-.',linewidth=3)
plt.title("Line Plot")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.grid(True)
plt.show
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.subplot(2,1,1)
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(x,y2,color='b',linestyle=':',linewidth=2)
plt.grid(True)
plt.show()
```

```python
student = {"Ammu":100, "Hemanth":90,"Kannu":10}
```

```python
names = list(student.keys())
values = list(student.values())
```

```python
plt.bar(names,values)
plt.show()
```

```python
plt.bar(names,values)
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(names,values,color="g")
plt.title(" Horinzontal Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
from matplotlib import pyplot as plt
x = [10, 20, 30, 40, 50, 60, 70, 80, 90]
a = [8,1,7,2,0,3,7,3,2]
plt.scatter(x,a)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 80, 90]
a = [8,1,7,2,0,3,7,3,2]
b = [7,2,5,6,9,1,4,5,3]
plt.scatter(x,a,marker="*",c="g",s=100)
plt.scatter(x,b,marker=".",c="r",s=200)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 80, 90]
a = [8,1,7,2,0,3,7,3,2]
b = [7,2,5,6,9,1,4,5,3]

plt.subplot(1,2,1)
plt.scatter(x,a,marker="*",c="g",s=100)

plt.subplot(1,2,2)
plt.scatter(x,b,marker=".",c="r",s=200)
plt.show()
```

```python
from sklearn.datasets import load_iris
import pandas as pd

data = load_iris()
df = pd.DataFrame(data.data, columns=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"])
df['Species'] = data.target

df.to_csv("Iris.csv", index=False)
```

```python
import pandas as pd
Iris = pd.read_csv('Iris.csv')
Iris.head()
plt.hist(Iris['SepalLengthCm'],bins=30,color="r")
plt.show()
```
